# Data Visualization using Seaborn

## The Data

We will be working with a titanic data set. 

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.jointplot(x = 'fare', y = 'age', data = titanic, kind='scatter')
```

```python
sns.distplot(titanic['fare'],bins=30,kde=False,color='red')
```

```python
sns.boxplot(x='class', y='age', data = titanic, palette='rainbow')
```

```python
sns.swarmplot(x='class', y='age', data=titanic, palette='inferno')
```

```python
sns.countplot(x='sex', data = titanic, palette='coolwarm')
```

```python
sns.heatmap(titanic.corr(), cmap='coolwarm')
plt.title('titanic.corr')
```

```python
a = sns.FacetGrid(data = titanic, col = 'sex')
a.map(plt.hist, 'age')
```
